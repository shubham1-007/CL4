### Installation:

1. Follow the installation steps we discussed earlier to install MongoDB on your Ubuntu system.

2. Once MongoDB is installed and running, you can proceed with the next steps.

### Creation of Database and Collection:

1. Open a terminal window.

2. Access the MongoDB shell by typing:
   ```bash
   mongo
   ```

3. To create a new database, use the `use` command followed by the name of the database you want to create. For example:
   ```bash
   use mydatabase
   ```
   Replace `mydatabase` with the name you want for your database.

4. Now, let's create a collection within the database. Collections are analogous to tables in relational databases. You can create a collection using the `db.createCollection()` method. For example:
   ```javascript
   db.createCollection("mycollection")
   ```
   Replace `mycollection` with the name you want for your collection.

### CRUD Operations:

#### Insert Document:
To insert a document into the collection, you can use the `insertOne()` or `insertMany()` method. Here's an example using `insertOne()`:
```javascript
db.mycollection.insertOne({ name: "John", age: 30, city: "New York" })
```

#### Query Document:
To query documents from the collection, you can use the `find()` method. Here's an example to find all documents in the collection:
```javascript
db.mycollection.find()
```
You can also add query criteria to filter documents. For example, to find documents where the age is greater than 25:
```javascript
db.mycollection.find({ age: { $gt: 25 } })
```

#### Update Document:
To update a document in the collection, you can use the `updateOne()` or `updateMany()` method. Here's an example using `updateOne()` to update a document matching a specific condition:
```javascript
db.mycollection.updateOne({ name: "John" }, { $set: { age: 35 } })
```

#### Delete Document:
To delete a document from the collection, you can use the `deleteOne()` or `deleteMany()` method. Here's an example using `deleteOne()` to delete a document matching a specific condition:
```javascript
db.mycollection.deleteOne({ name: "John" })
```
